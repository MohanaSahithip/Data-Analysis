```python
articles=[]
for i in range(len(links)):
    file_name = str(inputlinks.index[i])
    with open(file_name,'r') as file:
        articles.append(file.read())
        
```

```python
for i in range(len(articles)):
  articles[i]= articles[i].replace('\n','')
```

```python
sentences = []
for article in articles:
  sentences.append(len(sent_tokenize(article)))
```

```python
cleaned_articles = [' ']*len(articles)
```

```python
for i in range(len(articles)):
  for w in stop_words:
    cleaned_articles[i]= articles[i].replace(' '+w+' ',' ').replace('?','').replace('.','').replace(',','').replace('!','')
    #? ! , .
```

```python
words = []
for article in articles:
  words.append(len(word_tokenize(article)))
```

```python
words_cleaned = []
for article in cleaned_articles:
  words_cleaned.append(len(word_tokenize(article)))
```

```python
positive_score = [0]*len(articles)
for i in range(len(articles)):
  for word in pos_words:
    for letter in cleaned_articles[i].lower().split(' '):
      if letter==word:
        positive_score[i]+=1
```

```python
negative_score = [0]*len(articles)
for i in range(len(articles)):
  for word in neg_words:
    for letter in cleaned_articles[i].upper().split(' '):
      if letter==word:
        negative_score[i]+=1
```

```python
words_cleaned = np.array(words_cleaned)
sentences = np.array(sentences)
```

```python
df["positive_score"]=positive_score
df["negative_score"]=negative_score
```

```python
df['polarity_score'] = (df['positive_score']-df['negative_score'])/ ((df['positive_score'] +df['negative_score']) + 0.000001)
df['subjectivity_score'] = (df['positive_score'] + df['negative_score'])/( (words_cleaned) + 0.000001)
df['average_sentence_length'] = np.array(words)/np.array(sentences)
```

```python
complex_words = []
sylabble_counts = []
for article in articles:
  sylabble_count=0
  d=article.split()
  ans=0
  for word in d:
    count=0
    for i in range(len(word)):
      if(word[i]=='a' or word[i]=='e' or word[i] =='i' or word[i] == 'o' or word[i] == 'u'):
           count+=1
#            print(words[i])
      if(i==len(word)-2 and (word[i]=='e' and word[i+1]=='d')):
        count-=1;
      if(i==len(word)-2 and (word[i]=='e' and word[i]=='s')):
        count-=1;
    sylabble_count+=count    
    if(count>2):
        ans+=1
  sylabble_counts.append(sylabble_count)
  complex_words.append(ans)   
```

```python
df['percentage_of_complex_words'] = np.array(complex_words)/np.array(words)
df['fog_index'] = 0.4 * (df['average_sentence_length'] + df['percentage_of_complex_words'])
df['average_no_words_per_sentence'] = df['average_sentence_length']
df['complex_word_count'] = complex_words
df['word_count'] = words
df['syllable_per_word'] = np.array(sylabble_counts)/np.array(words)
```

```python
total_characters = []
for article in articles:
  characters = 0
  for word in article.split():
    characters+=len(word)
  total_characters.append(characters) 
```

```python
personal_nouns = []
personal_noun =['I', 'we','my', 'ours','and' 'us','My','We','Ours','Us','And'] 
for article in articles:
  ans=0
  for word in article:
    if word in personal_noun:
      ans+=1
  personal_nouns.append(ans)
df['personal_pronouns'] = personal_nouns
```

```python
df['average_word_length'] = np.array(total_characters)/np.array(words)
```
